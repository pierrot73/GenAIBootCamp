Exercice 1 : Understanding Data Visualization

Data visualization is crucial because it transforms complex data sets into visual formats like charts and graphs, making it easier to identify patterns, trends, and outliers. It helps analysts and stakeholders quickly grasp insights without sifting through raw numbers. Visualization enhances understanding, facilitates communication of findings, supports decision-making, and can reveal relationships that might be overlooked in traditional data tables.

Purpose of a Line Graph in Data Visualization:
A line graph is used to display data points connected by straight lines, primarily to show trends or changes over time. It is ideal for illustrating how a variable evolves, comparing multiple data series, and identifying patterns such as increases, decreases, or cyclical behaviors across intervals. Line graphs provide a clear visual representation of data progression, making them valuable for time-series analysis and trend identification.

```python
# Importation de la bibliothèque matplotlib.pyplot pour créer des graphiques
import matplotlib.pyplot as plt

# Liste des températures pour chaque jour de la semaine sur l'axe des y
temperatures = [72, 74, 76, 80, 82, 78, 75]

# Liste des jours de la semaine sur l'axe des x
jours = ['Lundi', 'Mardi', 'Mercredi', 'Jeudi', 'Vendredi', 'Samedi', 'Dimanche']

# Créer le graphique en traçant les températures en fonction des jours
plt.plot(jours, temperatures, marker='o')  # marker='o' pour marquer chaque point

# Ajouter un titre au graphique
plt.title('Variation de la température sur une semaine')

# Étiqueter l'axe des x
plt.xlabel('Jour')

# Étiqueter l'axe des y
plt.ylabel('Température (°F)')

# Afficher le graphique
plt.show()
```

```python
# Importer matplotlib.pyplot pour la gestion des graphiques
import matplotlib.pyplot as plt

# Liste des ventes mensuelles (exemple) en dollars
monthly_sales = [5000, 5500, 6200, 7000, 7500, 8000, 8200, 7800, 7300, 6800, 6500, 6000]

# Noms des mois pour l'axe des x
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Création du graphique à barres avec les ventes par mois
plt.bar(months, monthly_sales)

# Ajout du titre au graphique
plt.title('Monthly Store Sales')

# Étiquette pour l'axe des x
plt.xlabel('Month')

# Étiquette pour l'axe des y
plt.ylabel('Sales Amount ($)')

# Affichage du graphique
plt.show()
```

```python
# Importation de la bibliothèque pandas pour la manipulation des données
import pandas as pd
from google.colab import drive

# Lecture du csv
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/W2/D4/Student Mental health.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
# Importer seaborn pour la visualisation avancée
import seaborn as sns
# Importer matplotlib.pyplot pour la gestion des graphiques
import matplotlib.pyplot as plt

# Créer une nouvelle colonne pour indiquer si l'étudiant a de l'anxiété (Yes/No)
# Convertir en valeurs binaires pour faciliter la visualisation
df['Anxiety_binary'] = df['Do you have Anxiety?'].map({'Yes': 1, 'No': 0})

# Calculer la proportion d'étudiants ayant de l'anxiété pour chaque genre
# On groupe par genre puis on calcule la moyenne de 'Anxiety_binary'
axiety_proportion = df.groupby('Choose your gender')['Anxiety_binary'].mean().reset_index()

# Créer un bar plot pour comparer la proportion d'anxiété par genre
sns.barplot(data=axiety_proportion, x='Choose your gender', y='Anxiety_binary', palette='pastel')

# Ajouter un titre au graphique
plt.title('Proportion of Students Experiencing Anxiety by Gender')

# Afficher le graphique
plt.show()
```

```python
# Importation de la bibliothèque pandas pour la manipulation des données
import pandas as pd
from google.colab import drive
# Lecture du csv
drive.mount('/content/drive')
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/W2/D4/Student Mental health.csv')
df.head()
```

```python
# Importer seaborn pour la visualisation avancée
import seaborn as sns
# Importer matplotlib.pyplot pour la gestion des graphiques
import matplotlib.pyplot as plt

# Convertir la colonne 'Do you have Panic Attack?' en valeurs numériques (Yes=1, No=0)
df['Panic_Attack_Numeric'] = df['Do you have Panic attack?'].map({'Yes': 1, 'No': 0})

# Créer un histogramme de la distribution de l'âge en fonction de la présence d'attaques de panique
# La couleur peut différencier les groupes
sns.histplot(data=df, x='Age', hue='Panic_Attack_Numeric', multiple='stack', palette='Set2')

# Ajout d'un titre pour clarifier le graphique
plt.title('Distribution of Age by Panic Attack Occurrence')

# Ajouter des labels pour plus de clarté
plt.xlabel('Age')
plt.ylabel('Count')

# Afficher le graphique
plt.show()
```
