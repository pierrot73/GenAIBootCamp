```python
#Import Libraries
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from scipy.ndimage import rotate
import numpy as np

# Load an image using PIL
image_path = '/content/drive/MyDrive/Colab Notebooks/W2/D2/0026.png'
original_image = Image.open(image_path)

# Display the original image using matplotlib
plt.imshow(original_image)

# Fonction pour faire pivoter une image de 90 degrés
def rotate_image_90_degrees(image):
    # Rotation de 90 degrés dans le sens horaire
    return rotate(image, angle=90, reshape=True)
rotated_image=rotate_image_90_degrees(original_image)
plt.imshow(original_image)
```

```python
from PIL import Image, ImageOps
# Flipper une image horizontalement (miroir gauche/droite)
# puis verticalement (miroir haut/bas)
def flip_image_horizontal_et_vertical(image):
    # Flipper horizontalement
    image_flipped_horizontal = ImageOps.mirror(image)
    # Flipper verticalement
    image_flipped_vertical = ImageOps.flip(image_flipped_horizontal)
    return image_flipped_vertical

flipped_image = flip_image_horizontal_et_vertical(original_image)
plt.imshow(flipped_image)
```

```python
# Flipper horizontalement (miroir gauche/droite)
horizontal_flip = ImageOps.mirror(original_image)

plt.imshow(horizontal_flip)
```

```python
# Flipper verticalement (miroir haut/bas)
vertical_flip = ImageOps.flip(horizontal_flip)
plt.imshow(vertical_flip)
```

```python
from PIL import Image, ImageOps

# Zoomer sur une image en la redimensionnant (scale de 1.2x)
def zoom_in_image(image):
    # Obtenir la taille originale
    width, height = image.size
    # Calculer la nouvelle taille en la multipliant par 1.2
    new_size = (int(width * 10), int(height * 10))
    # Redimensionner l'image
    zoomed_image = image.resize(new_size, Image.LANCZOS)
    return zoomed_image

# zoomed_image = zoom_in_image(image)
zoomed_image = zoom_in_image(original_image)
plt.imshow(zoomed_image)
```
