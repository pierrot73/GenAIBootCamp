```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA

# Création du DataFrame
df = pd.DataFrame(data)

# 1. Normalisation Min-Max de la colonne 'salary'
scaler = MinMaxScaler()
df['salary_normalized'] = scaler.fit_transform(df[['salary']])

# 2. Supposons qu'il y ait d'autres features numériques pour PCA
# Pour l'exemple, créons des features artificielles
np.random.seed(0)
df['feature1'] = np.random.rand(len(df))
df['feature2'] = np.random.rand(len(df))
features = ['salary_normalized', 'feature1', 'feature2']

# 3. Application de PCA pour réduire à 2 composants
pca = PCA(n_components=2)
principal_components = pca.fit_transform(df[features])
df_pca = pd.DataFrame(data=principal_components, columns=['PC1', 'PC2'])

# Ajout des composants PCA au DataFrame original (optionnel)
df = pd.concat([df, df_pca], axis=1)

# 4. Groupement par 'experience_level' et calcul de la moyenne et médiane du salaire
grouped_stats = df.groupby('experience_level')['salary'].agg(['mean', 'median']).reset_index()

# Affichage des résultats
print("DataFrame avec salaires normalisés et PCA :")
print(df[['salary', 'salary_normalized', 'PC1', 'PC2']])
print("\nMoyenne et médiane des salaires par niveau d'expérience :")
print(grouped_stats)
```
