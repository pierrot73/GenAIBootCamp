Exercice 5

```python
import matplotlib.pyplot as plt
plt.plot (df['SepalLengthCm'])
plt.show()
import seaborn as sns

# 1. Calcul des statistiques de tendance centrale pour la longueur du sépale
mean_sepal_length = df['SepalLengthCm'].mean()
median_sepal_length = df['SepalLengthCm'].median()
mode_sepal_length = df['SepalLengthCm'].mode()[0]

print(f"Mean Sepal Length: {mean_sepal_length}")
print(f"Median Sepal Length: {median_sepal_length}")
print(f"Mode Sepal Length: {mode_sepal_length}")

# 2. Création d'un histogramme pour visualiser la distribution des longueurs du sépale
plt.figure(figsize=(8, 5))
sns.histplot(df['SepalLengthCm'], bins=5, kde=True)
plt.title('Distribution des longueurs du sépale')
plt.xlabel('Longueur du sépale (cm)')
plt.ylabel('Fréquence')
plt.show()
```
