```python
# üåü Exercice 1 : Hello World
# Affiche "Hello world" quatre fois sur une seule ligne.
print("Hello world\nHello world\nHello world\nHello world")


# üåü Exercice 2 : Some Math
# Calcule (99^3) * 8, c'est-√†-dire 99 √† la puissance 3, multipli√© par 8.
resultat = (99 ** 3) * 8
print(resultat)


# üåü Exercice 3 : Quel est ton nom ?
# Demande √† l'utilisateur son nom et compare avec le mien, puis affiche un message amusant.
mon_nom = "Alice"  # Par exemple, mon nom est Alice
nom_utilisateur = input("Quel est ton nom ? ")

if nom_utilisateur == mon_nom:
    print("Wow, on a le m√™me nom ! C'est la classe !")
else:
    print(f"Ton nom est {nom_utilisateur}, mais moi c'est {mon_nom}.")


# üåü Exercice 4 : As-tu la taille pour faire un roller coaster ?
# Demande la taille en cm et affiche un message selon la taille.
taille = int(input("Quelle est ta taille en centim√®tres ? "))

if taille > 145:
    print("Tu es assez grand(e) pour faire un tour de montagnes russes !")
else:
    print("Tu dois encore grandir un peu pour faire un tour de montagnes russes.")


# üåü Exercice 5 : Num√©ros favoris
# Travaillons avec des sets.
my_fav_numbers = {3, 7, 21}  # Mes num√©ros pr√©f√©r√©s
# Ajoutons deux nouveaux num√©ros
my_fav_numbers.add(42)
my_fav_numbers.add(99)

# Supprimer le dernier num√©ro ajout√© (le dernier qui a √©t√© ajout√© n'a pas d'ordre, mais on peut supprimer un sp√©cifique)
# Par exemple, supprimer 99
my_fav_numbers.remove(99)

# Cr√©er un set pour les num√©ros pr√©f√©r√©s de mon ami
friend_fav_numbers = {7, 11, 42}

# Concat√©ner (unionner) les deux sets pour cr√©er un nouveau set
our_fav_numbers = my_fav_numbers.union(friend_fav_numbers)

print("Nos num√©ros pr√©f√©r√©s : ", our_fav_numbers)


# üåü Exercice 6 : Tuple
# Les tuples sont immuables, on ne peut pas y ajouter directement.
mon_tuple = (1, 2, 3)

# Essayer d'ajouter un √©l√©ment provoquera une erreur
# On peut, par contre, cr√©er un nouveau tuple en concat√©nant
nouveau_tuple = mon_tuple + (4,)
print("Tuple initial :", mon_tuple)
print("Nouveau tuple :", nouveau_tuple)


# üåü Exercice 7 : Manipulation de listes
basket = ["Banana", "Apples", "Oranges", "Blueberries"]

# Supprimer "Banana"
basket.remove("Banana")
# Supprimer "Blueberries"
basket.remove("Blueberries")
# Ajouter "Kiwi" √† la fin
basket.append("Kiwi")
# Ajouter "Apples" au d√©but
basket.insert(0, "Apples")
# Compter combien de fois "Apples" appara√Æt
nombre_apples = basket.count("Apples")
print("Nombre de 'Apples' dans la liste :", nombre_apples)
# Vider la liste
basket.clear()
# Afficher la liste finale
print("Liste finale :", basket)


# üåü Exercice 8 : Commandes de sandwich
sandwich_orders = ["Tuna sandwich", "Pastrami sandwich", "Avocado sandwich", "Pastrami sandwich", "Egg sandwich", "Chicken sandwich", "Pastrami sandwich"]
# Enlever toutes les occurrences de "Pastrami sandwich"
while "Pastrami sandwich" in sandwich_orders:
    sandwich_orders.remove("Pastrami sandwich")

# Liste pour les sandwiches pr√©par√©s
finished_sandwiches = []

# Pr√©parer chaque sandwich
while sandwich_orders:
    current_sandwich = sandwich_orders.pop(0)  # retire le premier de la liste
    finished_sandwiches.append(current_sandwich)  # l'ajoute √† la liste termin√©e

# Afficher ce qui a √©t√© pr√©par√©
for sandwich in finished_sandwiches:
    print(f"J'ai pr√©par√© votre {sandwich.lower()}")
```
