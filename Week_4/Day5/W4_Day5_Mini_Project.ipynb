```python
# Importation des bibliothèques nécessaires
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder

# 1. Analyse exploratoire (EDA)

# Afficher les dimensions du dataset
print(f"Nombre de lignes : {df.shape[0]}, Nombre de colonnes : {df.shape[1]}")

# Visualisations (exemples)

# Distribution de l'âge
plt.figure(figsize=(8, 4))
sns.histplot(df['age'], bins=20, kde=True)
plt.title('Distribution de l\'âge')
plt.xlabel('Âge')
plt.ylabel('Nombre de patients')
plt.show()

# Répartition de la variable cible (présence de maladie cardiaque)
plt.figure(figsize=(6, 4))
sns.countplot(x='num', data=df)
plt.title('Répartition de la variable cible (maladie cardiaque)')
plt.xlabel('Présence de maladie (0 = non, 1 = oui)')
plt.ylabel('Nombre de patients')
plt.show()

# Analyse des corrélations entre variables numériques
# Sélectionner uniquement les colonnes numériques
numeric_cols = df.select_dtypes(include='number').columns

# Calculer la matrice de corrélation sur les colonnes numériques
corr = df[numeric_cols].corr()
plt.figure(figsize=(12, 8))
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Matrice de corrélation')
plt.show()

# 2. Prétraitement des données

# Encodage des variables catégoriques
# Identifier les colonnes catégoriques
categorical_cols = df.select_dtypes(include='object').columns
print("\nColonnes catégoriques :", categorical_cols)

# Si la colonne 'sex' par exemple est catégorique, on peut l'encoder
# Cependant, dans ce dataset, 'sex' est probablement numérique (0,1)
# Si besoin, on peut utiliser LabelEncoder
# Exemple :
# le encodeur
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Vérification de l'encodage
print("\nDonnées après encodage des catégoriques :")
print(df.head())

# Normalisation / Standardisation des variables numériques
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include='float64').columns
# Si 'id' est une colonne d'identifiant, on ne la scale pas
numeric_cols = [col for col in numeric_cols if col != 'id']

# Appliquer la normalisation
df[numeric_cols] = scaler.fit_transform(df[numeric_cols])

# Vérification après normalisation
print("\nDonnées après normalisation :")
print(df.head())

# 3. Sélection des features et de la variable cible
X = df.drop('num', axis=1)
y = df['num']

# À ce stade, le dataset est prêt pour la modélisation
print("\nFeatures (X) et variable cible (y) préparées pour la modélisation.")
```

```python
import pandas as pd

# Vérifier s'il y a des NaN dans l'ensemble du dataset
nan_presence = df.isnull().any().any()

if nan_presence:
    print("Il y a des valeurs NaN dans le dataset.")
    # Optionnel : afficher le nombre de NaN par colonne
    print("Nombre de NaN par colonne :")
    print(df.isnull().sum())
else:
    print("Il n'y a pas de valeurs NaN dans le dataset.")
```

Interprétations :

Pour les variables numériques comme "trestbps", "chol", "oldpeak", "ca", on peut utiliser l'imputation par la moyenne ou la médiane.
Pour des variables comme "ca" (nombre de vaisseaux), il pourrait être pertinent de faire une imputation spécifique ou de considérer leur distribution.
La variable "id" n'est probablement pas utile pour la modélisation, donc je la supprime.
La variable "num" étant la cible, je ne la modifie pas.

J'utilise la variable cible pour une imputation spécifique :
Pour être plus précis je fais une imputation conditionnelle en utilisant la moyenne ou la médiane par classe de la variable cible pour la variable 'Chol'. Parce que les valeurs manquantes pourraient dépendre de la présence ou absence de la maladie, ainsi cette stratégie respecte la relation entre les variables.

Recommandation générale :
Si la variable "num" est la cible, j'évite d'utiliser ses valeurs pour l'imputation dans cette variable pour ne pas biaiser le modèle. En revanche, pour d’autres variables, faire une imputation en utilisant la médiane/mode globales ou par sous-groupes en fonction de "num" peut améliorer la qualité des données.

```python
import pandas as pd

# Suppression de la colonne 'id'
df.drop('id', axis=1, inplace=True)

# Imputation des NaN pour les variables numériques
# Pour 'trestbps' (tension artérielle au repos)
df['trestbps'].fillna(df['trestbps'].median(), inplace=True)

# Imputation conditionnelle pour 'chol' par classe de 'num'
for target_value in df['num'].unique():
    mask = df['num'] == target_value
    median_chol = df.loc[mask, 'chol'].median()
    df.loc[mask & df['chol'].isnull(), 'chol'] = median_chol

# Imputation conditionnelle pour "thalch" de la même façon que 'chol'
median_thalch = df['thalch'].median()
df['thalch'].fillna(median_thalch, inplace=True)

# Pour 'oldpeak' (ancien pic d'ischémie)
df['oldpeak'].fillna(df['oldpeak'].median(), inplace=True)

# Pour 'ca' (nombre de vaisseaux principaux)
# Étant un entier, on peut imputer par la médiane ou la mode
df['ca'].fillna(df['ca'].mode()[0], inplace=True)

# Vérification
print("NaN après traitement :")
print(df.isnull().sum())
```

2. Model Training:

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.impute import SimpleImputer

# --- 1. Vérification initiale ---
print("NaNs avant imputation :")
print(X.isnull().sum())

# --- 2. Imputation de toutes les colonnes numériques ---
imputer = SimpleImputer(strategy='median')

# Appliquer l'imputer à toutes les colonnes de X
X_imputed = pd.DataFrame(imputer.fit_transform(X), columns=X.columns)

# --- 3. Vérification après imputation ---
print("NaNs après imputation :")
print(X_imputed.isnull().sum())

# Si des NaN persistent, cela indique un problème dans la DataFrame
# (par exemple, NaN dans la cible ou NaN dans des colonnes non numériques)

# --- 4. Séparer les données ---
X_train, X_test, y_train, y_test = train_test_split(
    X_imputed, y, test_size=0.3, random_state=42, stratify=y)

# --- 5. Entraîner le modèle ---
model = LogisticRegression(max_iter=1000, random_state=42)
model.fit(X_train, y_train)

# --- 6. Prédictions et évaluation ---
y_pred = model.predict(X_test)

print("Rapport de classification :")
print(classification_report(y_test, y_pred))
print("Matrice de confusion :")
print(confusion_matrix(y_test, y_pred))
```

3. Model Evaluation:

```python
from sklearn.metrics import accuracy_score, precision_recall_fscore_support, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Prédictions sur l'ensemble de test ---
y_pred = model.predict(X_test)

# --- 2. Calcul de l'accuracy ---
accuracy = accuracy_score(y_test, y_pred)
print(f"Exactitude (accuracy) : {accuracy:.2f}")

# --- 3. Calcul des autres métriques (précision, rappel, F1) ---
# average='weighted' pour tenir compte du déséquilibre entre classes
precision, recall, f1, _ = precision_recall_fscore_support(y_test, y_pred, average='weighted')

print(f"Précision (precision) : {precision:.2f}")
print(f"Rappel (recall) : {recall:.2f}")
print(f"F1 score : {f1:.2f}")

# --- 4. Visualisation de la matrice de confusion ---
cm = confusion_matrix(y_test, y_pred)

# Création d'une heatmap pour visualiser la matrice
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=range(cm.shape[1]), yticklabels=range(cm.shape[0]))
plt.xlabel('Prédictions')
plt.ylabel('Véritables valeurs')
plt.title('Matrice de confusion')
plt.show()
```

### Interprétation des résultats d’évaluation du modèle

- **Exactitude (accuracy) : 61%**  
  Cela signifie que le modèle prédit correctement la présence ou l’absence de maladie cardiaque pour environ 61% des patients du jeu de test. C’est une performance modérée, indiquant qu’il peut faire des prédictions correctes plus de la moitié du temps.

- **Précision (precision) : 0.57**  
  La précision indique la proportion de patients prédits comme ayant une maladie cardiaque qui en ont réellement une. Ici, lorsque le modèle prédit la maladie, il a environ 57% de chances d’être correct. Cela montre une tendance à certains faux positifs.

- **Rappel (recall) : 0.61**  
  Le rappel indique la capacité du modèle à détecter tous les patients réellement atteints de la maladie. Avec un rappel de 61%, le modèle identifie une majorité des patients malades, mais il en manque encore près de 39% (faux négatifs).

- **F1 score : 0.59**  
  C’est une moyenne harmonique entre précision et rappel, donnant une idée globale de la performance, équilibrant faux positifs et faux négatifs. Ici, le score montre une performance modérée.

- **Matrice de confusion** :  
  La matrice montre comment les prédictions se répartissent par rapport aux véritables classes. On peut voir que le modèle a tendance à confondre certaines classes, notamment celles proches en termes de caractéristiques médicales, ce qui est attendu dans un problème multi-classe avec des classes déséquilibrées.
